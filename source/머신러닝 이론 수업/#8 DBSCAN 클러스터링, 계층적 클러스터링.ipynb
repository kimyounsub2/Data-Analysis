# DBSCAN 클러스터링
## DBSCAN 클러스터링 개요
### DBSCAN이란?
* Density-based spatial clustering of applications with noise 의 약어
* k평균 클러스터링과 같이 데이터의 위치 정보 이용
* 데이터의 밀도를 활용하여 클러스터링
    * k평균 클러스터링은 평균관의 거리가 얼마나 떨어졌는지 결정
    * 같은 클러스터 내의 데이터들은 밀도가 높을 것이라는 가정

### DBSCAN 파라미터 정의
* 첫번째, 주변 공간에 대한 정의
* 두번쨰, 그 주변 공간에 몇 개의 데이터가 존재해야 클러스터로 설정할 것인가?

### DBSCAN의 클러스터링
* 파라미터들을 활용하여 클러스터링 하거나 노이즈를 제거하는 알고리즘
* 클러스터는 핵심 벡터를 중심으로 생성
* 핵심 벡터에서 접근은 가능하지만 핵심 벡터가 아닌 외곽 벡터 존재

### DBSCAN의 장점
* 클러스터의 개수를 미리 정할 필요가 없음
* 각각의 데이터들에 해해 밀도를 계산해서 클러스터링
* 비선형 경계의 클러스터링 가능
* 노이즈 데이터를 따로 분류하므로 노이즈 값이 클러스터에 영향을 주지 않음
* 클러스터에서의 경계에서 애매하게 있는 점이 줄어듦
    * DBSCAN의 N(최소 이웃의 수) 개수 때문

### DBSCAN의 단점
* 데이터를 사용하는 순서에 따라 클러스터링이 다르게 될 수 있음
    * 2개 이상의 각기 다른 군집에 핵심 벡터가 있을 경우, 이 순서에 따라 외곽 벡터가 다른 클러스터에 속할 수 있음
    * 자주 발생하는 문제는 아님
    * 발생하여도 클러스터에 큰 영향을 주는 문제가 아님
* 거리측정 방법에 따라 고차원 데이터에서 적절한 E를 찾기 어려움
* DBSCAN 클러스터링의 수행절차
    * 반경 E와 최소 개수 N의 정의 -> 핵심 벡터 선정 -> 외곽 벡터 선정 -> 노이즈 선정 -> 클러스터와 노이즈 분류
    
### R에서 DBSCAN을 계산하기 위한 명령어
* dbscan()

# 계층적 클러스터링
## 계층적 클러스터링 개요
### 계층적 클러스터링이란?
* 고차원 또는 다차원 데이터를 시각화 하는 기본적 방법 중 하나
* 사용이 매우 간단함
* 사람이 확인하기 쉬움
    * 방법이 직관적임
    * 고차원의 데이터 셋에서 어떤 일이 일어나는지 빠르게 확인
* 계층적 군집 이라고도 불림
* 데이터들을 '계층화'하는 작업

### 클러스터 접근 절차
* 흩어져 있는 작은 조각들이 모임 -> 작은 공 형태로 무리가 형성 -> 형성된 무리는 다른 무리와 합체

### 클러스터 접근 절차의 체크 포인트
* 거리 측정 기준으로 사용하는 두 점 사이의 거리를 어떻게 계산할까?
* 점끼리 어떻게 합칠 것인가?
    * 주로, 덴드로그램의 형태로 나타남

### 거리 측정 방법
* 유클리디안 거리 측정 : 두 점 사이의 직선 거리
* 맨하탄 거리 : 각 블록을 이동하듯이 두 점 사이의 거리 측정

### 계층적 클러스터링 수행 절차
* N개의 자료를 하나의 클러스터로 취급
* 주어진 클러스터들 중 가장 가까운 두 클러스터를 병합
* 병합 과정을 반복하여 하나의 클러스터에 속할 때까지 반복
* 거리 츨정 방법에 따라 최단 연결, 최장 연결로 구분
