# 머신러닝 적용
## 머신러닝 적용 프로세스
### 머신러닝 적용 대상
* 논리적 분류 : 명확한 조건에 의한 분류
* 머신러닝 : 학습한 경험에 의해 스스로 분류 조건이 개선됨

### 머신러닝 적용 방법
* 머신러닝 절차 : 데이터 수집 -> 데이터 전처리 -> 데이터 학습 -> 모델 평가
* 데이터 수집
    * 해결하려고 하는 문제 확인
    * 문제의 답을 위한 데이터 정의
    * 데이터는 어디에서 얻을 것인가?
    * 공공데이터 활용, 직접 수집 등
* 데이터 전처리
    * 누락 값 정리 : 누락 데이터 삭제 또는 추정해서 채우기
    * 특이점 제거 : 신뢰 구간을 벗어난 값 제거
    * 데이터 변환 : 표준화, 정규화, 이산화 등
    * 데이터 축소 : 예측력이 떨어질 수 있는 데이터 줄이기
* 데이터 학습
    * 문제의 종류와 수집된 데이터의 유형에 따라 적절한 학습 방법과 알고리즘을 선택
    * 지도학습, 자율학습, 강화학습
    * 회귀분석, 분류, 클러스터링
    * 서포트 벡터 머신, 의사 결정 트리, 나이브 베이즈 분류 등
* 모델 평가
    * 학습 데이터와 검증 데이터 구분
    * 교차 평가

## 머신러닝 적용 사례
### 자율주행 자동차
* 정해진 트랙 위에서의 자동 운행
    * 기차의 경우, 정해진 선로로만 통행함
    * 자동 운전이 일부 가능
    * 안전 조건만 만족하면 정해진 선로에서 자동운전 하므로 논리적 판단으로 가능
* 자율주행 자동차
    * 자동차 도로의 명확한 조건을 결정하기 어려움(차선이 지워진 곳 등)
    * 모든 도로를 기차 선로처럼 구축 하는 것이 불가능함
    * 자동차 운전은 사람의 경험에 의해 이루어짐
    * 다양한 돌발 상황은 항상 새롭게 발견되므로 스스로 학습하며 운전이 이루어져야 함
    
### 네이버 랩스
* 국토부의 임시주행 허가를 받음
* 2017년 서울 모터쇼에서 자율주행 자동차 전시
* 2017년 현재 SAE(미국 자동차공학회 자율주행 기술 수준) 레벨 3 수준으로 개발 완료
    * SAE 레벨 3 : 부분 자율주행 단계(운전자 조작 없이 목적지까지 일부 운행)
    * SAE 레벨 4 : 정해진 조건에서 운전자 개입 없이 차량의 속도와 방향 통제

### 구글 자율주행 자동차
* SAE 레벨 4 수준
* 90% 이상 스스로 운전 가능
* 자율주행 자동차를 전담 개발하는 웨이모
* 2015년 12월 부터 2016년 11월까지 102만 km의 주행 경험 보유

### 농업에서의 활용
* 구글의 머신러닝 기술인 텐서플로우(TensoFlow)활용
* 오이 분류기 개발
* 카메라를 이용하여 오이의 사진을 찍고 오이 선별
* 텐서플로우에 오이 사진을 2~3일 정도 7,000장으로 학습
* 완벽하지는 않지만 오이가 학습을 통해 분류

### 공공기관에서의 활용
* 코리아텍 직업능력심사평가원에서 직업훈련 부정수급 차단
    * 직업능력심사평가원 주 업무는 정부 지원금을 훈련기관에 지원하고 감시함
    * 훈련기간 7천여 곳, 5만여 개 훈련과정을 관리해야 함
    * 정부 예산 부정 수급 방법은 지능화 되어 감
    * 직업훈련정보망의 빅데이터를 활용
    * 데이터 마이닝, 시나리오 분석 기법으로 부정 패턴 확인
    * 머신러닝을 통한 부정 수급 의심 사항을 찾아내어 감시함

### 정보보안에서의 활용
* 악성코드 탐지
    * 소프트웨어에 악성코드를 담아서 배포되는 경우가 많음
    * 악성코드 방지를 위해 지정된 공간에서만 실행되는 샌드박스(Sanbox) 방법을 많이 활용함
    * 제한된 환경의 샌드박스 방법은 불편하고 한계점이 많음
    * 머신러닝을 통해 정상적인 프로그램과 악성코드 탑재 프로그램을 학습
    * 지속적인 학습을 통해 자동으로 악성코드를 분류하도록 활용
