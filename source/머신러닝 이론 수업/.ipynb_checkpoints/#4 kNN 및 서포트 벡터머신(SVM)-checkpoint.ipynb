# kNN
## kNN 알고리즘 소개
### kNN이란?
* kNN(k nearest neighbors) : 최근접 이웃 알고리즘
* 새로운 데이터의 분류를 알기 위해 사용
* k값
    * 기존의 데이터와 새로운 데이터를 비교
        * 새로운 데이터와 가장 인접한 데이터 k개를 선정
        * k개의 데이터가 가장 많이 속해 있는 분류를 선택
    * k 값에 의해 결정된 분류를 새로운 데이터의 분류로 확정
### kNN의 특징
* 장점
    * 정확도가 높음
    * 오류 데이터가 결과값에 크게 영향을 미치지 않음
    * 데이터에 대한 가정이 없음
* 단점
    * 느림
    * 많은 메모리를 사용함
    
## kNN 알고리즘 활용 사례
### 합기술 특허 문서의 IPC 분류 예측
* IPC는 그 특징에 따라 다양한 코드가 부여됨, 새로운 특허에 대해 IPC를 어떻게 부여할지 결정이 필요함
* kNN을 활용하여 새로운 특허의 IPC분류 예측
    * 기존의 IPC가 부여된 특허를 분석함
    * 새로운 특허와 유사한 분야의 기존 특허를 검출
    * 검출된 유사한 상위 k개의 특허의 분류를 확인
    * 상위 k개가 속한 특허의 분류를 새로운 특허에 적용함
* 총 6,233개의 문서 분류 예측
    * 1,123개의 문서는 100% 정확히 예측함
    * 많은 수의 문서가 50% 이상 예측함
### 고속도로 통행시간 예측
* 현재의 실시간 교통 정보의 문제점
    * 현재 시간의 정보는 알 수 있지만, 출발 후 향후 교통상황 예측이 불가함
* 교통예보 지원시스템의 문제점
    * 기존 자료만 활용
    * 실시간 자료를 반영하지 않으므로, 통행패턴이 과거와 상이하면 정확도가 낮음
* 사용자의 요구
    * 현재 출발하여 도착지까지 어느정도 시간이 걸릴지 실시간 예측을 원함
* 연구방법
    * 1단계 : 관련 연구 사례 검토
    * 2단계 : 서울TG ~ 대전IC 구간의 TCS와 DSRC자료를 이용하며, kNN모형 구축
    * 3단계 : 예측 모형 정확도 검증
* 서울 TG -> 대전IC 간의 통행량 예측 오차율 비교
* 기존의 방법(ARIMA,Exponential Smoothing)에 비해 kNN 방법이 오차율이 낮음
    * 기존의 방법은 실시간 특성을 반영하지 않음
    * kNN은 기존의 데이터와 실시간 데이터를 비교하여 예측하여 오차가 낮아짐
    * kNN을 위한 기존의 데이터가 더욱 많이 축적되면 오차율이 더 낮아질 수 있음
    

# 벡터머신(SVM)
## SVM 알고리즘 소개
### SVM이란?
* Support Vector Machine 의 약자
* Support  Vector 와 Hyperplane(초평면)이 주요 개념
* 신경망에 비하여 간결함
* 분류(Classification)이나 회귀 분석(Regression)에서 사용가능
* 분류에서 주로 사용
* 기본적으로 Hyperplane을 이용해서 분류를 함

### 선형분류
* 평면 상태에서 기준에 의해 분류함
* 초평면에 의해 분류가 가능한 상태

### Support Vector을 통한 분류
* 최고의 마진을 가져가는 방법으로 분류 기준을 세움
* 비어 있는 마진이 많아야 새로운 데이터가 들어와도 분류가 잘 될 가능성이 큼

### SVM의 단점
* 여러 개의 조합 테스트가 필요함
* 학습 속도가 느려 질 수 있음
* 해석이 어렵고 복잡한 블랙박스 형태로 됨

### 데이터 특성에 따른 분류
* 오류가 일부 발생하더라도 마진을 최대한 넓게 하는 방법
* 오류가 절대 발생하지 않게 하는 방법

### SVM의 일반화 적용
* 마진이 넓은 것을 최우선(데이터 일반화)
* 오류 발생 가능성을 최소로 함(데이터 일반화를 하지 않음)

## 비선형 특징 분류
### 비선형 데이터란?
* 초평면으로 데이터가 일정하게 나누어져 있지 않은 경우
* 하나의 지건 형태로 단순하게 분류가 불가능함

### 위상 변화를 통한 초평면 분류
* 데이터의 위상을 변화하여 초평면에 의해 분류가 가능하도록 함
* 구별이 가능한 방향으로 사상(mapping)을 시키면 새로운 공간 영역으로 변환
* 새로운 공간 영역에서는 초평면에 의해 분류가 가능해짐

### Mapping을 통한 선형 구별
* X, Y 기준의 2차원 평면 상에서는 선형으로 구분이 불가능한 데이터의 변환
* Mapping 함수를 활용하여 3차원 평면으로 변경

### 커널트릭
* 저차원에서 선형 구분이 불가능한 데이터를 고차원으로 변경해야 함
* 고차원으로 변경된 데이터를 선형적으로 구분
* 활용할 수 있는연산은 내적 연산만 가능

### 커널 함수
* 대표적으로 다항커널(Polynomial Kernel)과 가우시안 커널(Gaussian Kernel)이 있음
