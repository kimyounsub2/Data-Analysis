# 데이터 샘플링 및 R프로그램 활용
## 데이터 샘플링
### 데이터 샘플링이란?
* 데이터를 일부 정리해서 최적의 입력 데이터로 만드는 과정

### 확률적 샘플링 : 순수하게 무작위(Random)로 샘플링하는 방법
* 단순 랜덤 샘플링 : 전체 데이터에서 무작위로 샘플을 추출하는 방법
* 2단계 샘플링 : m개의 하위 모집 단 중에서 n개의 데이터를 랜덤으로 추출 
* 층별 샘플링 : 각 층에서 랜덤하게 데이터 추출
* 군집, 집락 샘플링 : 군집 중, 하나 또는 여러 개의 군집을 선정
* 계통 샘플링 : 일정 간격마다 하나씩 데이터 추출

### 비확률적 샘플링
* 사전에 표본으로 선정되는 확률을 모르는 상태에서 주관적으로 추출
* 샘플링하는 사람의 주관적인 의도가 개입되어 있는 샘플링
* 암시적 모집단 : 비확률적 샘플링에서 나타나는 표본의 모집단
* 이상적 모집단 : 주제에 관련해 가장 적합한 모집단
    * 암시적 모집단과 이상적 모집단은 일치해야함
* 편의 샘플링 : 시점 및 위치 등이 진행에 편리하게 표본을 선택
* 판단 샘플링 : 목적에 가장 적합한 표본이라고 생각하는 대상을 선택하는 방법
* 할당 샘플링 : 모집단을 세그먼트로 구분하여 각 세그먼트에 표본 수를 나타내는 쿼타를 할당

## R프로그램 활용
### R이란?
* 통계분석과 그래픽을 작성하기 위한 프로그래밍 언어
    * 통계학자를 위한 프로그래밍 언어
* 분석 소프트웨어 : 수많은 알고리즘과 방법론 제공
    * 데이터 입출력
    * 데이터 처리
    * 데이터 분석
    * 그래프 작성
    
## R의 특징
* 언제 어디서나 무료로 사용가능
* in-memory Computing : 빠른 처리속도, 메모리의 크기에 따라 성능 영향
* OOP(Objext Oriented Programming) : 객체지향, 클래스와 메소드로 구성
* 패키지 : 다양한 알고리즘이나 방법론이 패키지 형태로 제공됨


# 지도학습 개요
## 지도학습 모델 소개
### 지도학습이란?
* 컴퓨터에게 어떤 것이 맞는 답인지를 지정해 줌
* 목적 값(Target Value)이 있음
* 컴퓨터는 지정해 준 답과 비슷한 것으 ㄹ판단해서 맞는 것이 무엇인지 판단함
* 판단을 하기 위해 수 많은 데이터를 활용하여 학습함
* 지도학습 방법으로는 예측(Regression)과 분류(Classification)이 있음

### 지도학습 방법
* 목적 값의 형태에 따라 구분함
* 분류(Classification) : 목적 값의 연솏성이 없이 몇 가지로 구분됨
* 예측(Regression) : 목적 값의 연속성이 있음

## R을 활용한 지도학습 모델 실습
### 데이터 로딩 : R프로그램에서 사용할 수 있도록 수집된 데이터 로딩 및 플로팅
### 회귀분석 : 데이터를 회귀분석 하여 그래프로 확인함
* 훈련 집합에 대한 학습을 기반으로 테스트 집합에 대해 적확한 분류 또는 값을 할당하는 함수를 찾는것

### 분석 결과 요약 : 회귀분석 결과를 통해 데이터 분석
* Call : 회귀분석에 사용된 모델 식
* Residuals : 잔차, 즉 회귀선의 값과 실제 관측 값의 차이를 각 분위수로 표시
* Coefficients : 절편, 독립변수 등에 대한 회귀계수를 표시
* Residual standard error : 잔차의 표준오차와 자유도를 나타냄
* Multiple R-squared : 결정계수, 즉 추정된 회귀선이 실제 관측괎을 얼마나 잘 설명하는가를 나타냄
* Adjusted R-squared : 수정결정계수, 주로 다중회귀분석에서 독립변수를 증가시키면 결정계수가 올라가는 것에 대한 대안으로 사용함
* F-statistic : F통계량은 해당 모형이 의미가 있는지 아닌지를 알려줌
