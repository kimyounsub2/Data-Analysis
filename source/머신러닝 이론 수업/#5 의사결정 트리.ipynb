# 의사결정 트리
## 의사결정트리 개요
### 의사결정트리 정의
* 과거에 수집된 데이터들을 분석하여 이들 사이에 존재하는 패턴(범주별 특성)을 속성의 조합으로 나타내는 분류 모형
* 새로운 데이터에 대한 분류(Classification)
* 해당 범주의 값을 예측(Prediction)
* 데이터로부터 트리 구조의 일반화된 지식을 추출

### 의사결정트리 구조
* 대표적으로 노드(node), 가지(branch), 깊이(depth)로 구성

### 의사결정트리 장점
* 이해하기 쉬운 규칙의 생성
* 분류예측에 유용
* 연속형, 범주형 변수 모두 취급 가능
* 어느 변수가 상대적으로 더 중요한지를 보여줌
* 비교적 빠른 속도로 의사결정이 가능함

### 의사결정트리 단점
* 연속형 변수값을 예측 할 때 적당하지 않음
* 복잡한 트리 모형
    * 예측력이 저하됨
    * 해석이 어려움
    * 상황에 따라 계산량이 많아짐
* 비 안정성
    * 데이터에 약간의 변형이 있는 경우 결과가 나빠질 수 있음

## 의사결정 방법
### 의사결정 진행절차
* 의사결정트리 모형 구축
* 의사결정트리 분리
* 반복적 분리 과정
* 분리 기준 : 이산형 목표 변수, 연속형 목표 변수
* 불순도 측정
    * 지니 지수 : 지니 지수를 감소시켜주는 예측변수와 그 때의 최적분리에 의해서 자식마디를 선택
    * 엔트로피 지수 : 무질서도에 대한 측정, 엔트로피 지수가 가장 작은 예측 변수와 이 때의 최적분리에 의해 자식마디를 형성
* 정지 기준 설정

### 의사결정 알고리즘
* CART
    * 가장 널리 사용되는 의사결정트리 알고리즘
    * 불순도의 측도로서 출력변수가 범주형인 경우 지니 지수를 이용
    * 출력변수가 연속형인 경우에는 분산을 이용하여 이진분리(binary split)
    * 입력변수들의 선형결합 중에서 최적의 분리를 찾을 수 있음
    
* C4.5와 C5.0
    * CART와는 다르게 각 마디에서 다자분리(multiple split)가 가능
    * 범주형 입력변수에 대해서는 범주의 수만큼 분리가 일어남
    * 불순도의 측도로는 엔트로피지수를 사용
    
* CHAID
    * 가지치기를 하지 않고 적당한 크기에서 나무모형의 성장을 중시
    * 입력변수가 반드시 범주형 변수이어야 함
    * 불순도의 측도로는 카이제곱 통계량을 사용
    * 변수들 간의 통계적 관계를 찾는 것이 그 목적
    * 분류기법(Classification Technique)으로써 사용할 수 있음
