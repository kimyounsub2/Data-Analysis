# 과접합과 일반화
## 과접합의 의미
### 과접합이란 만들어진 모델이 지나치게 복잡하여 학습 데이터에 대해서는 올바른 결과를 출력하지만 실제 데이터가 입력 되었을 때 올바른 결과를 출력하지 못하는 것을 말함

### 과접합의 원인
* 데이터 편향
* 데이터 오류

## 일반화의 의미
### 실제 데이터가 학습 데이터와 차이가 있더라도 생성된 모델의 성능 차이가 나지 않게 하는 것을 일반화라고 함
### 일반화의 방법
* 속성 줄이기
* 정규화
* 검증

# 데이터 분할
## 데이터 준비
### 데이터 준비가 필요한 이유
* 다양한 환경에서 수집되는 데이터는 다양한 문제점이 발생함
    * 불환전 데이터
    * 손상이 많은 데이터
    * 일치하는 값이 없는 데이터
* 데이터의 품질이 나쁘면 머신러닝 수행 결과도 나빠짐
* 머신러닝 결과를 통해 우수한 예측 시스템을 만들려면 그 데이터가 우수해야 함

### 좋은 데이터란?
* 머신러닝을 수행할 목적에 맞는 데이터
    * 예) 알파고를 만들려고 하는데 축구 경기 룰 데이터가 필요할 것인가
* 손상되지 않은 데이터
* 데이터에 충돌하는 레코드가 없는데이터
* 데이터에 일치하지 않는 값이 없는 데이터
* 머신러닝을 수행하기에 충분한 양의 데이터

### 데이터 상태 검사 방법
#### 항목 : 설명
* 레코드 수 : 데이터 레코드의 개수
* 기능 수 : 데이터 레코드의 개수
* 데이터 유형 : 데이터가 연속적 또는 특정한 값
* 누락된 값의 수 : 누락된 데이터 값의 수
* 데이터 형식 : 데이터가 XML이라면 그 형태에 맞는지?
* 일관되지 않은 데이터 레코드 : 값의 허용범위(0~5 값만 사용하는데 그 이외가 입력되는지?)

### 데이터 준비 절차의 주요 작업
* 데이터 정리
* 데이터 변환
* 데이터 감소
* 데이터 분할
* 데이터 문자 정리

## 데이터 분할 절차
### 누락 값 처리
* 삭제 : 값이 누락 된 레코드를 제거
* 더미 대체 : 누락된 값을 더미 값으로 대체
* 평균 대체 : 누락된 값이 숫자이면 평균으로 대체
* 빈도 대체 : 누락된 값이 범주이면 가장 빈도가 높은 항목으로 대체
* 회귀 대체 : 회귀 함수를 사용하여 누락된 값을 대체

### 데이터 분할
* 연속 값을 조건에 맞게 데이터를 분할함
* 특성에 따른 분할
    * 예) 해당 값이 속하는 범주에 따른 분할
* 간격에 따른 분할
    * 예) 1~9, 10~19, 20~29와 같이 특정 간격에 따른 데이터 분할
