# 상관분석
### 상관분석의 개념
* 상관분석 : 두 변수 간에 관계가 있는지를 알아보고자 할 때 실시하는 분석방법
* 상관관계 : 두 변수(대상)이 서로 관련성이 있다고 추측되는 관계
* 광관계수 : 상관분석에서 두 변수의 관련된 정도를 나타내주는 값

### 상관분석의 특징
* 양의 상관 : 하나의 변수가 커질수록 다른 변수도 함께 커지는 경우
* 음의 상관 : 하나의 변수가 커질수록 다른 변수는 오히려 작아지는 경우
* 무상관 : 변수끼리 서로 상관이 없는 경우
* 상관계수의 값 : -1 <= r <= 1
* 두변수의 관련성이 클수록 상관계수는 1에 가까워지고 두 변수의 관련성이 적을수록 상관계수는 0에 가까워짐

### R을 이용한 상관분석
* 상관분석 그래프를 구현하기 위해 'corrplot'와 lattice 패키지를 설치함
* cor 함수를 이용하여 상관 계수를 산출함
* ggplot2 패키지를 이용하여 데이터 시각화

# 회귀분석
### 회귀분석의 개념
* 회귀 : 한 바퀴 돌아 제자리로 돌아오거나 돌아간다는 뜻으로, 19세기 영국의 우생학자 프랜시스 캘턴이 키가 큰 부모의 자식들의 키가 점점 더 커지지 않고 평균 키로 회귀하는 경향을 발견하면서 쓰임
* 회귀분석 : 독립변수가 종속변수에 영향을 미치는지 확인하기 위해 실시하는 분석 방법으로 독립 변수와 종속변수 사이에 인과관계가 존재할 대, 그 관계의 정도를 분석하고 통게적 유의성을 검증하는 것
* 최소제곱법 : 관측점들과 회귀선 간의 수직 거리(잔차)를 제곱하여 각각 더한 값 각각의 관측값들에서 추정된 직선까지의 거리 제곱합이 최소가 되도록 회귀계수를 구하는 것

### 회귀분석의 특성
* 회귀방정식 : 회귀분석에서 독립변수가 종혹변수에 미치는 영향에 대해 나타내는 것
* 회귀계수 : 회귀계수는 변수의 관련성에 따라 양수(+) 혹은 음수(-)의 형태를 띔
* 회귀분석의 구분
* 구분 요인 / 분석방법
    * 독립변수의 수 : 1개 - 단순회귀분석 , 2개 이상 - 다중 회귀분석
    * 독립변수의 척도 : 명목/서열 척도 - 더미변수 회귀분석, 등간/비율 척도 - 일반 회귀분석
    * 독립변수와 종속변수의 관계 : 선형 - 선형 회귀분석, 비선형(연속적) - 비선형 회귀분석
* 다중회귀분석 : 둘 이상의 독립변수가 종속변수에 미치는 영향에 대해 나타내는 경우
    * 둘 이상의 독립변수가 종속변수에 미치는 영향에 대해 나타내는 경우
    * 도출된 회귀선이 변수간 관계를 얼마나 잘 설명하는지는 회귀선과 데이터와의 적합도 수준을 이용하여 평가됨
    * 이 때, 설명력의 지표로는 상관계수, 회귀선의 기울기, 결정계수(R2), 수정된 R2이 사용됨
* 회귀분석과 상관분석
    * 차이점 : 회귀분석은 변수 간의 인과관계를 검증하고, 상관분석은 변수 간의 관계 여부를 검증하는 분석볍
    * 공통점 : 변수간의 관계성을 검증

### R을 이용한 회귀분석
* 근무년수가 연봉에 영향을 미치는가에 대한 분석을 학습함
* cor() 함수를 이용하여 상관계수를 추출함
* lm() 함수를 이용하여 회귀분석을 실시함
* 회귀분석 실시 후 나오 p-value 값으로 귀무가설 기각 여부 결정함


# 분산분석
### 분산분석의 개념
* 두 집단 간 속성에 대한 평균 차이를 검증하는 방법으로 사용하는 t검정의 비효율성을 줄이기 위해 집단 간 변화량과 집단내 변화량을 비교하는 방법으로 사용
* F분포에 근거하여 검정
* 두 개 이상 집단들의 평균을 비교하는 통계분석 기법
* 집단들의 평균 차이를 비교하기 위해 사용

### 분산분석의 특성
* 분산분석의 기본 가정 : 정규성, 분산동일성, 독립성
* F통계량(F-value) : 집단간 분산과 집단내 분산의 비
* 일원분산분석 : 요인(집단을 구분하는 독립변수)이 하나인 경우, 모집단의 수에 제한이 없음, 각 표본의 수가 같지 않아도 됨
* 이원분산분석 : 요인(집단을 구분하는 독립변수)이 둘인 경우, 요인이 결과에 미치는 영향을 알아보기 위한 주효과와 상호작용 효과를 살펴볼 수 있음
* 다원배치 분산분석 : 독립변수가 둘 이상인 경우를 총칭

### R을 이용한 분산분석
* 거식증 환자의 치료방법에 따른 몸무게의 변화가 있는가 가설 설정
* 데이터 입력(R제공 anorexia 데이터 사용)
* anova 함수를 사용해 분산분석을 실시함
* oneway 함수를 사용해 분산분석을 실시함
* P값이 0.05보다 작으므로, 귀무가설을 기각
