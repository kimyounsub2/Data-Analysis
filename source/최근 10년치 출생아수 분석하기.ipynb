```python
import pandas as pd 
```

```python
df_kosis = pd.read_csv('../data/data/출생아수_시도_시_군_구.csv', engine='python')
df_kosis.shape
```

```python
pd.options.display.max_columns = 865
df_kosis.head(n=1)
```

```python
df = df_kosis.melt(id_vars='시군구별')
df.head()
```

```python
df['시군구별'].unique()
```

```python
print(df.shape)
df = df[df['시군구별'] != '시군구별'].copy()
df.shape
```

```python
df['시군구별'].unique()
```

```python
df.head()
```

```python
df['연도'] = df['variable'].str.split('.', expand=True)[0]
df['월'] = df['variable'].str.split('.', expand=True)[1]
df['성별'] = df['variable'].str.split('.', expand=True)[2]
df.head()
```

```python
df.tail()
```

```python
df['성별'].unique()
```

```python
df['성별'] = df['성별'].fillna('전체')
df['성별'].unique()
```

```python
df['성별'] = df['성별'].replace('1','남자').replace('2','여자')
df['성별'].unique()
```

```python
df['성별'].value_counts()
```

```python
df.head()
```

```python
df = df.rename(columns={'variable':'기간', 'value':'출생아수'})
df.head()
```

```python
df.info()
```

```python
import numpy as np
df['출생아수'] = df['출생아수'].replace('-', np.nan)
```

```python
df['출생아수'] =  df['출생아수'].astype(float)
```

```python
df['출생아수'].describe()
```

```python
df_all = df[(df['시군구별'] =='전국') & (df['성별'] == '전체')]
df_all.head(n=2)
```

```python
df_all = df_all[['연도','월','출생아수']].copy()
df_all.head()
```
